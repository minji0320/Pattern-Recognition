# Taylor Approximation

##### 1. Define a differentiable function that maps from real number to real number.
$$f(x) = 7x^3+2x^2+3x+5$$

##### 2. Define a domain of the function.
$$f:X \longrightarrow Y$$

$$X = \{x| -100 \le x \le 100\}$$

##### 3. Plot the function.

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 7*x**3 + 2*x**2 + 3*x + 5

x = np.arange(-100,100)
plt.plot(x,f(x),label='f(x)')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

##### 4. Select a point within the domain.
$$P = (50,\ f(50)) = (50,\ 880,155)$$

##### 5. Mark the selected point on the function.

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 7*x**3 + 2*x**2 + 3*x + 5

x = np.arange(-100,100)
plt.plot(x,f(x),label='f(x)')
plt.plot(50,f(50),'ro',label='P')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

##### 6. Define the first-order Taylor approximation at the selected point.
$$f(x) = 7x^3+2x^2+3x+5 \quad at\ 50$$
$$f'(x) = 21x^2+4x+3$$

$$\begin{matrix}
\hat{f}(x) &=& f(50) + f'(50)(x-50)\\
            &=& 880,155 + 52,703(x-50)\\
            &=& 52,703x - 1,754,995
\end{matrix}$$

##### 7. Plot the Taylor approximation with the same domain of the original function.

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 7*x**3 + 2*x**2 + 3*x + 5
def taylor(x):
    return 52703*x - 1754995
x = np.arange(-100,100)
plt.plot(x,f(x),label='f(x)')
plt.plot(50,f(50),'ro',label='P')
plt.plot(x,taylor(x),'r--',label='Taylor')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

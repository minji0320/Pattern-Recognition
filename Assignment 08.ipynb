# Polynomial Fitting

Solve a least square problem to find an optimal polynomial curve for a given set of two dimensional points.

Demonstrate the effect of the degree of polynomial in fitting a given set of points.

## Choose a polynomial curve and generate points along the curve with random noise

Choose a polynomial curve :
$$f(x) = \sin \frac{x}{100} \times 700 + \cos \frac{x}{30} \times 300 + x \quad (-500 \le x < 500)$$

Generate points along the curve with random noise : 
$$g(x) = f(x) + \eta \quad  (-50 < \eta < 50)$$


```python
import numpy as np
from numpy.linalg import *
import matplotlib.pyplot as plt
import random


np.random.seed(1)

def f(x):
    return np.sin(x/100)*700 + np.cos(x/30)*300 + x

def g(x):
    return f(x) + (np.random.normal(size=len(x)))*100

x = np.arange(-500,500)
```

## Plot the generated noisy points along with its original polynomial without noise

```python
plt.plot(x,f(x),label= 'f(x)')
plt.scatter(x, g(x), marker='.' ,color='red', label = 'g(x)')

plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

## Plot the approximating polynomial curve obtained by solving a least square problem

Define a polynomial function and degree:
$$pf(x) = \theta_0 + \theta_1 x + \theta_2 x^2 +  \theta_3 x^3$$

```python
degree = 3

A = np.ones((1000,degree+1))
for i in range(1,degree+1):
    A[:,i] = x**i
    
theta = inv(A.T@A)@A.T@g(x)

for i in range(degree+1):
    print('theta',i,'=',theta[i])

def pf(x):
    return A@theta

plt.plot(x,f(x),label= 'f(x)')
plt.scatter(x, g(x), marker='.' ,color='red', label = 'g(x)')
plt.plot(x,pf(x),label= 'pf(x)')

plt.title('Approximating Polynomial Curve (degree = 3)')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
